```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw=pd.read_csv('train.csv')
```

```python
raw.head()
```

```python
raw.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
raw.isnull().sum()
```

```python
raw.Age.fillna(raw.Age.mean(),inplace=True)
```

```python
raw.Cabin.fillna('U',inplace=True)
```

```python
raw['Cabin']=raw.Cabin.astype(str).str[0]
```

```python
raw.Embarked.fillna('S',inplace=True)
```

```python
raw['Sex']=raw.Sex.apply(lambda x: 0 if x=='male' else 1)
```

```python
cabin_dummies=pd.get_dummies(raw.Cabin,prefix='Cabin',drop_first=True)
```

```python
cabin_dummies
```

```python
raw=pd.concat([raw,cabin_dummies],axis=1)
```

```python
raw.drop('Cabin',axis=1,inplace=True)
```

```python
raw['Embarked']=raw.Embarked.apply(lambda x: 0 if x=='S' else (1 if x=='C' else 2))
```

```python
raw.Embarked.value_counts()
```

```python
raw.columns
```

```python
raw=raw[[ 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare',
       'Embarked', 'Cabin_B', 'Cabin_C', 'Cabin_D', 'Cabin_E', 'Cabin_F',
       'Cabin_G', 'Cabin_T', 'Cabin_U','Survived']]
```

```python
raw['Age']=raw['Age'].apply(lambda x: x/raw['Age'].max())
```

```python
raw['Fare']=raw['Fare'].apply(lambda x: x/raw['Fare'].max())
```

```python
matrix=np.array(raw.values,'float')
```

```python
x=matrix[:,:-1]
y=matrix[:,-1].reshape(-1,1)
```

```python
x
```

```python
y
```

```python
x=np.concatenate((x,np.ones((len(x),1))),axis=1)
```

```python
thetha0=np.zeros((x.shape[1],1))+0.5
```

```python
thetha0
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
def find_cost(h,y):
    return -1/len(x)*np.sum((y * np.log(h) + (1 - y) * np.log(1 - h)))
def gradient(h):
    return np.dot(x.T, (h - y)) / y.size
def predict_probs(X, theta):
    return sigmoid(np.dot(X, theta))
def predict(X, theta, threshold=0.5):
    return predict_probs(X, theta) >= threshold
```

```python
def GradientDescent(thetha0 , learning_rate = 0.2 , epochs=10000 , TOL=0.00000002):
    
    thetha_history = [thetha0]
    z = np.dot(x, thetha0)
    h = sigmoid(z)
    cost_history = [find_cost(h,y)]
   
    print(f'epoch \t Cost \t')
    for epoch in range(epochs):
        print(f'{epoch:5d}\t{cost_history[-1]:7.4f}\t')
        z = np.dot(x, thetha0)
        h = sigmoid(z)
        grad = gradient(h)
        cost = find_cost(h,y)
        
        thethanew = thetha0 - learning_rate*grad
        thetha_history.append(thethanew)
        cost_history.append(cost)
        
        if np.sum((thethanew - thetha0)**2) < TOL:
            print('Convergence achieved.')
            break
        thetha0 = thethanew

    return thethanew , thetha_history , cost_history
```

```python
thetha , thetha_history , cost_history = GradientDescent(thetha0)
```

```python
predict(x,thetha)
```
